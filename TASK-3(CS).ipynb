```python
credit_card_data=pd.read_csv('/content/creditcard.csv')
```

Handling Null values

```python
credit_card_data['V7'].fillna(credit_card_data['V7'].mode()[0],inplace=True)
credit_card_data['V8'].fillna(credit_card_data['V8'].mode()[0],inplace=True)
credit_card_data['V9'].fillna(credit_card_data['V9'].mode()[0],inplace=True)
credit_card_data['V10'].fillna(credit_card_data['V10'].mode()[0],inplace=True)
credit_card_data['V11'].fillna(credit_card_data['V11'].mode()[0],inplace=True)
credit_card_data['V12'].fillna(credit_card_data['V12'].mode()[0],inplace=True)
credit_card_data['V13'].fillna(credit_card_data['V13'].mode()[0],inplace=True)
credit_card_data['V14'].fillna(credit_card_data['V14'].mode()[0],inplace=True)
credit_card_data['V15'].fillna(credit_card_data['V15'].mode()[0],inplace=True)
credit_card_data['V16'].fillna(credit_card_data['V16'].mode()[0],inplace=True)
credit_card_data['V17'].fillna(credit_card_data['V17'].mode()[0],inplace=True)
credit_card_data['V18'].fillna(credit_card_data['V18'].mode()[0],inplace=True)
credit_card_data['V19'].fillna(credit_card_data['V19'].mode()[0],inplace=True)
credit_card_data['V20'].fillna(credit_card_data['V20'].mode()[0],inplace=True)
credit_card_data['V21'].fillna(credit_card_data['V21'].mode()[0],inplace=True)
credit_card_data['V22'].fillna(credit_card_data['V22'].mode()[0],inplace=True)
credit_card_data['V23'].fillna(credit_card_data['V23'].mode()[0],inplace=True)
credit_card_data['V24'].fillna(credit_card_data['V24'].mode()[0],inplace=True)
credit_card_data['V25'].fillna(credit_card_data['V25'].mode()[0],inplace=True)
credit_card_data['V26'].fillna(credit_card_data['V26'].mode()[0],inplace=True)
credit_card_data['V27'].fillna(credit_card_data['V27'].mode()[0],inplace=True)
credit_card_data['V28'].fillna(credit_card_data['V28'].mode()[0],inplace=True)
credit_card_data['Amount'].fillna(credit_card_data['Amount'].mode()[0],inplace=True)
credit_card_data['Class'].fillna(credit_card_data['Class'].mode()[0],inplace=True)
```

This dataset is highly unbalanced

0.0-->Normal Transaction

1.0-->fradulent Transaction


```python
# separating the data for analysis
legit=credit_card_data[credit_card_data.Class==0.0]
fraud=credit_card_data[credit_card_data.Class==1.0]
```

```python
print(legit.shape)
```

```python
print(fraud.shape)
```

Statistical measures of the data

```python
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
# compare the values for both transactions
credit_card_data.groupby('Class').mean()
```

Under-Sampling

Build a sample dataset containing similar distributions of normal transactions and fraudulent transactions

```python
# Number of Fraudulent Transaction -->103
legit_sample = legit.sample(n=103)
```

Concatenating two DataFrames

```python
new_dataset = pd.concat([legit_sample,fraud],axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

Splitting the data into Features and Targets

```python
x=new_dataset.drop(columns='Class',axis=1)
y=new_dataset['Class']
```

```python
print(x)
```

```python
print(y)
```

Splitting the data into training data and Testing data

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2, stratify=y,random_state=2)
```

```python
print(x.shape,x_train.shape,x_test.shape)
```

Model Training

Logistic Regression

```python
model=LogisticRegression()
```

```python
# training the Logistic Regression Model with training Data
model.fit(x_train,y_train)
```

Accuracy Score

```python
# for training data
x_train_prediction=model.predict(x_train)
training_data_accuracy=accuracy_score(x_train_prediction,y_train)
```

```python
print("Accuracy of the Training Data: ",training_data_accuracy)
```

```python
# for testing data
x_test_prediction=model.predict(x_test)
testing_data_accuracy=accuracy_score(x_test_prediction,y_test)
```

```python
print("Accuracy of the Testing Data: ",testing_data_accuracy)
```
